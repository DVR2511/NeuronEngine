## Comparative Analysis and Visualization of Modern Activation Functions

To understand and visualize ELU, SELU, Swish, GELU and Linear (Identity) activation functions using Python

**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm   # for GELU
```

**Define Activation Functions**

**Generate Input Values**

- It creates 400 evenly spaced numbers between -10 and 10.

 linspace means:

- It generates numbers between two values with equal spacing.

```python
x = np.linspace(-10, 10, 400)
```

### ðŸ”µ ELU (Exponential Linear Unit)
**Formula:** $$ \text{ELU}(x) = \begin{cases} x, & x > 0 \\ \alpha (e^x - 1), & x \leq 0 \end{cases} $$ **Behavior:**
 - Smooth curve for negative values
 - Reduces dead neuron problem
 - Mean output closer to zero
 - Faster convergence than ReLU
 - Slightly slower due to exponential computation
 - Output range: (-Î±, âˆž)

 ðŸ“Œ Used when ReLU underperform

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

y = elu(x)

plt.plot(x, y)
plt.title("ELU Activation Function")
plt.grid()
plt.show()
```

### ðŸŸ¢SELU (Scaled Exponential Linear Unit)
**Formula:** $$ \text{SELU}(x) = \lambda \begin{cases} x, & x > 0 \\ \alpha (e^x - 1), & x \leq 0 \end{cases}
$$ **Behavior:**

- Self-normalizing activation
- Maintains mean â‰ˆ 0 and variance â‰ˆ 1
- Works best with specific initialization
- Used in deep feedforward networks
- Output range: (-Î» Î±, âˆž)

ðŸ“Œ(Standard constants used in research)

```python
def selu(x):
    alpha = 1.67326324
    lam = 1.05070098
    return lam * np.where(x > 0, x, alpha * (np.exp(x) - 1))

y = selu(x)

plt.plot(x, y)
plt.title("Scaled Exponential Linear Unit")
plt.grid()
plt.show()
```

###ðŸŸ¡Swish Activation Function

**Formula:**

$$
\text{Swish}(x) = x \cdot \sigma(x)
$$

**Behavior:**

- Smooth and non-monotonic  
- Better performance than ReLU in deep models  
- Used in modern CNNs  
- Computationally heavier than ReLU  
- Output range: (-âˆž, âˆž)

ðŸ“Œ Used in EfficientNet

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)


y = swish(x)

plt.plot(x, y)
plt.title("Swish Activation Function")
plt.grid()
plt.show()
```

### ðŸŸ£GELU (Gaussian Error Linear Unit)

**Formula:**

$$
\text{GELU}(x) = x \cdot \Phi(x)
$$

*(Î¦ is the Gaussian CDF)*

**Behavior:**

- Probabilistic activation  
- Smooth alternative to ReLU  
- Strong performance in NLP models  
- Used in Transformers (BERT, GPT)  
- Output range: (-âˆž, âˆž)

ðŸ“Œ Very important for NLP

```python
def gelu(x):
    return x * norm.cdf(x)


y = gelu(x)

plt.plot(x, y)
plt.title("Gaussian Error Linear Unit")
plt.grid()
plt.show()
```

### âš«Linear (Identity) Activation Function

**Formula:**

$$
f(x) = x
$$

**Behavior:**

- No non-linearity  
- Used in regression output layers  
- Output range: (-âˆž, âˆž)


```python
def linear(x):
    return x


y = linear(x)

plt.plot(x, y)
plt.title("Linear (Identity) Activation Function")
plt.grid()
plt.show()
```

**Compute Outputs**

```python
y_elu = elu(x)
y_selu = selu(x)
y_swish = swish(x)
y_gelu = gelu(x)
y_linear = linear(x)
```

# Plot Activation Functions

```python
plt.figure(figsize=(10,6))

plt.plot(x, y_elu, label="ELU", linewidth=2)
plt.plot(x, y_selu, label="SELU", linewidth=2)
plt.plot(x, y_swish, label="Swish", linewidth=2)
plt.plot(x, y_gelu, label="GELU", linewidth=2)
plt.plot(x, y_linear, label="Linear", linewidth=2)

plt.axhline(0)
plt.axvline(0)

plt.title("Advanced Activation Functions")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.legend()
plt.grid(True)

plt.show()
```
