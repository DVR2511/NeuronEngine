# Understanding Activation Functions
To understand and visualize Sigmoid, Tanh, ReLU,Leaky ReLU and Softmax activation functions using Python


**Import Library**


```python
import numpy as np
import matplotlib.pyplot as plt
```

**Define activation functions**

### Sigmoid Activation Function

**Formula:**

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

**Behavior:**

- Flattens at extremes → vanishing gradient  
- Slow learning in deep networks  
- Used mainly in output layer (binary classification)  
- Smooth S-shaped curve  
- Converts input into probability-like values  
- Sigmoid outputs values between 0 and 1  


```python
# Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### Tanh (Hyperbolic Tangent) Activation Function

**Formula:**

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

**Behavior:**

- Zero-centered  
- Better gradient flow than Sigmoid  
- Still suffers from vanishing gradient  
- Used in RNNs / older neural networks  
- Strong gradients near zero  
- Tanh outputs values between -1 and 1  


```python
# Tanh
def tanh(x):
    return np.tanh(x)
```

### ReLU (Rectified Linear Unit) Activation Function

**Formula:**

$$
\text{ReLU}(x) = \max(0, x)
$$

**Behavior:**

- Outputs 0 for negative inputs  
- Linear for positive inputs  
- Very fast and computationally efficient  
- Helps reduce vanishing gradient problem  
- Most commonly used in hidden layers  
- Can suffer from dead neuron problem  
- Output range: [0, ∞)


```python
# ReLU
def relu(x):
    return np.maximum(0, x)
```

### Leaky ReLU Activation Function

**Formula:**

$$
\text{Leaky ReLU}(x) =
\begin{cases}
x, & x > 0 \\
\alpha x, & x \leq 0
\end{cases}
$$

**Behavior:**

- Allows small negative values  
- Fixes dead neuron problem of ReLU  
- Better gradient flow for negative inputs  
- Slightly slower than ReLU  
- Common value of α = 0.01  
- Output range: (-∞, ∞)


```python
# Leaky ReLU
def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)
```

**Softmax**

### Softmax Activation Function

**Formula:**

$$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$

**Behavior:**

- Converts raw scores into probabilities  
- Output values sum to 1  
- Used in multi-class classification output layer  
- Emphasizes largest values  
- Sensitive to large inputs  
- Output range: (0, 1)


```python
# Softmax
def softmax(x):
    exp_x = np.exp(x - np.max(x))   # for numerical stability
    return exp_x / np.sum(exp_x)
```

**Generate input values**

```python
x = np.linspace(-10, 10, 400)
```

**Compute outputs**

```python
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
```

**Plot Sigmoid, Tanh, ReLU, Leaky ReLU**

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y_sigmoid, label="Sigmoid", linewidth=2)
plt.plot(x, y_tanh, label="Tanh", linewidth=2)
plt.plot(x, y_relu, label="ReLU", linewidth=2)
plt.plot(x, y_leaky_relu, label="Leaky ReLU", linewidth=2)

plt.axhline(0)
plt.axvline(0)

plt.title("Activation Functions Visualization")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.legend()
plt.grid(True)

plt.show()
```

**Visualize Softmax** separately


Softmax works on vectors, not single values.


```python
# Example input vector
x_softmax = np.array([1.0, 2.0, 3.0, 4.0])
y_softmax = softmax(x_softmax)

# Plot Softmax
plt.figure(figsize=(6, 4))
plt.bar(range(len(y_softmax)), y_softmax)
plt.title("Softmax Output Probabilities")
plt.xlabel("Class Index")
plt.ylabel("Probability")
plt.show()
```
